```python
import requests
import pandas as pd
```

```python
response = requests.get('https://financialmodelingprep.com/api/v3/symbol/NASDAQ?apikey=1iijAmmSTdMwq9CRVj1G4qXTneMFBSqx').json()
```

```python
df = pd.DataFrame.from_records(response)
```

```python
df = df.iloc[:, [0, 1,2,3,11,12]]
```

```python
df_dropped = df.drop_duplicates(subset=['name'], keep='first')
```

```python
# INCOME STATEMENT
# PLEASE TEST ONCE WITH 'nasdaq_symbols_1' --> IF IT WORKS; PLEASE USE 'nasdaq_symbols'
```

```python
%%time
for index, symbol in enumerate(nasdaq_symbols):
    url = f'https://financialmodelingprep.com/api/v3/income-statement/{symbol}?period=quarter&apikey=1iijAmmSTdMwq9CRVj1G4qXTneMFBSqx'
    response = requests.get(url).json()
    df_response = pd.DataFrame.from_records(response)
    if index == 0:
        df_all = df_response
    else:
        df_all = pd.concat([df_all, df_response])
df_all.to_csv('raw_data/income_statement.csv')
```
